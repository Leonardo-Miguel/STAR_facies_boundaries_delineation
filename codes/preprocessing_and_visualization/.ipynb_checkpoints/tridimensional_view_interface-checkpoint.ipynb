This viewer was developed from the cigvis library (https://github.com/JintaoLee-Roger/cigvis), especially from the viserplot module, which supports visualization 3D data in web/browser environment.

The viewer has three overlapping plot options:

- data_path (str): Any volume (e.g., seismic)
- mask_path (str): A mask over that volume (e.g., label or RGT)
- tridimensional_plot_path (str): Any 3D surface (e.g., channel or horizon)

### Warning!

The data can be in NumPy array or sgy/segy format. In the case of segy data, the data opening parameters must be provided.
The 'iline' and 'xline' parameters are the fields where segy searches for information about the inlines and crosslines of the data.
The default value for segy data is iline=189 and xline=193, but Petrobras generally uses iline=21 and xline=25.

- iline_xline_data (tuple): (iline, xline)
- iline_xline_mask (tuple): (iline, xline)
- iline_xline_tridimensional_plot (tuple): (iline, xline)

Furthermore, you can choose the color map for each plot, which are the following parameters:

- color_map_data (str): Some matplotlib cmap
- color_map_mask (str): Some matplotlib cmap
- color_map_tridimensional_plot (str): Some matplotlib cmap

### Warning!

When you run the script, you'll see a link to access the visualization and interact with the plot. However, you'll need to replace localhost with the hostname you're using. You can query the hostname directly via the terminal, as in the example.

```python
from cigvis.visernodes import Server
from importlib import reload
import TridimensionalViewer
reload(TridimensionalViewer)

from TridimensionalViewer import ViserPlotIntegrated

def create_server(port, label='cigvis-viser', verbose=False):
    return Server(label=label, port=port, verbose=verbose)
    
##################################################################################
        
server = create_server(12345)

#'/datasets/public_real_data/parihaka_dataset.npy'

data_path = '/datasets/public_real_data/parihaka_dataset.npy'
color_map_data = None
iline_xline_data = (189, 193)

mask_path = None # '/pgeoprj/godeep/dados/l1_datasets/parihaka/facies.sgy'
color_map_mask = None
iline_xline_mask = (189, 193)

tridimensional_plot_path = '/datasets/public_real_data/parihaka_boundaries.npy' #'/datasets/models_predictions/parihaka_10.1a_boundaries_prediction.npy'
color_map_tridimensional_plot = None
iline_xline_tridimensional_plot = (21, 25)

##################################################################################

visualizer = ViserPlotIntegrated(
    server=server,
    data_path=data_path,
    iline_xline_data=iline_xline_data,
    color_map_data=color_map_data,
    mask_path=mask_path,
    iline_xline_mask=iline_xline_mask,
    color_map_mask=color_map_mask,
    tridimensional_plot_path=tridimensional_plot_path,
    iline_xline_tridimensional_plot=iline_xline_tridimensional_plot,
    color_map_tridimensional_plot=color_map_tridimensional_plot
)

visualizer.create_nodes()
```
