```python
def plot_steps(plot_data):

    plt.figure(figsize=(10,8), dpi=500)
                    
    plt.subplot(1,6,1)
    plt.imshow(plot_data[0].T, cmap='RdGy')
    plt.title('H × W')
    plt.axis("off")
    
    plt.subplot(1,6,2)
    plt.imshow(plot_data[1].T, cmap='RdGy')
    plt.title('H/2 × W/2')
    plt.axis("off")

    plt.subplot(1,6,3)
    plt.imshow(plot_data[2].T, cmap='RdGy')
    plt.title('H/4 × W/4')
    plt.axis("off")

    plt.subplot(1,6,4)
    plt.imshow(plot_data[3].T, cmap='RdGy')
    plt.title('H/8 × W/8')
    plt.axis("off")

    plt.subplot(1,6,5)
    plt.imshow(plot_data[4].T, cmap='RdGy')
    plt.title('H/16 × W/16')
    plt.axis("off")

    plt.subplot(1,6,6)
    plt.imshow(plot_data[5].T, cmap='RdGy')
    plt.title('H/32 × W/32')
    plt.axis("off")

    plt.tight_layout()
    plt.show()

def apply_padding(sections):

    def next_multiple_of_32(n):
        if n % 32 == 0:
            return n
        return (n // 32 + 1) * 32

    C, H, W = sections.shape

    H_pad = next_multiple_of_32(H)
    W_pad = next_multiple_of_32(W)

    pad_h = H_pad - H
    pad_w = W_pad - W

    if pad_h == 0 and pad_w == 0:
        return sections

    # padding por seção (aplicado igual para cada canal)
    padded = np.pad(
        sections,
        pad_width=((0, 0), (0, pad_h), (0, pad_w)),
        mode='constant'
    )

    return padded

def max_pool(sections):

    C, H, W = sections.shape
    block_height, block_width = 2, 2

    sec = sections.reshape(
        C,
        H // block_height, block_height,
        W // block_width, block_width
    )

    print(sec.shape)
    # (C, H/2, W/2, 4)
    sec_flat = sec.transpose(0, 1, 3, 2, 4).reshape(
        C, H // block_height, W // block_width, 4
    )

    # índice do maior em módulo
    idx = np.abs(sec_flat).argmax(axis=-1)

    # seleção
    out = np.take_along_axis(sec_flat, idx[..., None], axis=-1)[..., 0]

    return out
```

```python
data_file = '/datasets/public_real_data/parihaka_dataset.npy'

data = np.load(data_file, mmap_mode='r')
n_poolings = 5
samples = [230, 430]
#samples = [9, 51, 68, 80, 96, 107, 124, 136, 147, 172, 191, 206, 221, 235, 248, 258, 273, 285, 314, 336, 356, 385, 410, 428, 453, 474, 505, 528, 552, 579]

example_data = data[0:100, :, :]
#example_data = np.expand_dims(example_data, axis=0)
example_data = apply_padding(example_data)
print(example_data.shape)
plots_data = [example_data[99, :, :]]

data_pooled = example_data.copy()
for i in range(n_poolings):
    data_pooled = max_pool(data_pooled)
    section_to_plot = data_pooled[99, :, :]
    plots_data.append(section_to_plot)

print()
plot_steps(plots_data)
```
