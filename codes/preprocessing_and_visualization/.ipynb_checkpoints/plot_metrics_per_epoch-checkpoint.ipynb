```python
import matplotlib.pyplot as plt
import json
import glob
from matplotlib.lines import Line2D
from matplotlib.colors import ListedColormap
```

```python
def plot_metric(metrics_train_file, metrics_val_file):

    experiment_name = metrics_train_file.split('/')[-2]

    if 'SSL' in metrics_train_file:

        fig, ax = plt.subplots(1, 2, figsize=(8, 3))
        ax = ax.flatten()

        with open(metrics_train_file, 'r') as f:
            train_r2_data = json.load(f)[f'train_r2']
    
        with open(metrics_val_file, 'r') as f:
            val_r2_data = json.load(f)[f'val_r2']

        with open(metrics_train_file, 'r') as f:
            train_loss_data = json.load(f)[f'train_loss']
    
        with open(metrics_val_file, 'r') as f:
            val_loss_data = json.load(f)[f'val_loss']
            
        mean_r2 = [(a + b) / 2 for a, b in zip(train_r2_data, val_r2_data)]
        best_mean_r2 = max(mean_r2)
        print(best_mean_r2)
        idx_best_mean_r2 = mean_r2.index(best_mean_r2)

        ax[0].axvline(x=idx_best_mean_r2, color='r', linestyle='--', label='Max mean R2')
        ax[0].plot(train_loss_data, label=f'Loss train')
        ax[0].plot(val_loss_data, label=f'Loss validation')
        #ax[0].set_ylim(0, 0.06) #max(max(train_loss_data), max(val_loss_data)) + 0.1)
        ax[0].set_title('Loss')

        ax[1].axvline(x=idx_best_mean_r2, color='r', linestyle='--', label='Max mean R2')
        ax[1].plot(train_r2_data, label=f'R2 score train')
        ax[1].plot(val_r2_data, label=f'R2 score validation')
        ax[1].set_ylim(-0.25, 1)
        ax[1].set_title('R2')
        ax[1].legend()

    else:

        fig, ax = plt.subplots(1, 3, figsize=(12, 4))
        ax = ax.flatten()

        with open(metrics_train_file, 'r') as f:
            train_dist_data = json.load(f)[f'train_bdte']
    
        with open(metrics_val_file, 'r') as f:
            val_dist_data = json.load(f)[f'val_bdte']
            
        with open(metrics_train_file, 'r') as f:
            train_iou_data = json.load(f)[f'weighted_train_iou'] # [f'train_dte']
    
        with open(metrics_val_file, 'r') as f:
            val_iou_data = json.load(f)[f'weighted_val_iou'] # [f'val_dte']

        with open(metrics_train_file, 'r') as f:
            train_loss_data = json.load(f)[f'train_loss']
    
        with open(metrics_val_file, 'r') as f:
            val_loss_data = json.load(f)[f'val_loss']

        if len(val_dist_data) >= 99:
            min_val_dist = min(val_dist_data[99:])
        else:
            min_val_dist = min(val_dist_data)
        idx_min_dist = val_dist_data.index(min_val_dist)

        ax[0].axvline(x=idx_min_dist, color='r', linestyle='--')
        ax[0].plot(train_loss_data)
        ax[0].plot(val_loss_data)
        ax[0].set_title('Loss')
        #ax[0].set_ylim(150, 170) #0, max(val_loss_data) + 0.1)
        
        ax[1].axvline(x=idx_min_dist, color='r', linestyle='--')
        ax[1].plot(train_iou_data)
        ax[1].plot(val_iou_data)
        ax[1].set_ylim(0, 1)
        ax[1].set_title('WIoU')

        ax[2].axvline(x=idx_min_dist, color='r', linestyle='--', label='Min BDTE')
        ax[2].plot(train_dist_data, label='Train')
        ax[2].plot(val_dist_data, label='Validation')
        ax[2].set_ylim(0, 20)
        ax[2].legend()
        ax[2].set_title('BDTE')

    plt.tight_layout()
    plt.suptitle(experiment_name.capitalize(), y=1.05)
    plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.2)
    plt.legend(loc='upper left', bbox_to_anchor=(1.05, 0.6), borderaxespad=0.)
    plt.show()
```

```python
experiment_dir = '/datasets/models/parihaka_4a'
metrics_train_file = f'{experiment_dir}/all_metrics_train.json'
metrics_val_file = f'{experiment_dir}/all_metrics_validation.json'
plot_metric(metrics_train_file, metrics_val_file)

experiment_dir = '/datasets/models/parihaka_4b'
metrics_train_file = f'{experiment_dir}/all_metrics_train.json'
metrics_val_file = f'{experiment_dir}/all_metrics_validation.json'
plot_metric(metrics_train_file, metrics_val_file)

experiment_dir = '/datasets/models/parihaka_4c'
metrics_train_file = f'{experiment_dir}/all_metrics_train.json'
metrics_val_file = f'{experiment_dir}/all_metrics_validation.json'
plot_metric(metrics_train_file, metrics_val_file)

experiment_dir = '/datasets/models/parihaka_4d'
metrics_train_file = f'{experiment_dir}/all_metrics_train.json'
metrics_val_file = f'{experiment_dir}/all_metrics_validation.json'
plot_metric(metrics_train_file, metrics_val_file)

experiment_dir = '/datasets/models/parihaka_4e'
metrics_train_file = f'{experiment_dir}/all_metrics_train.json'
metrics_val_file = f'{experiment_dir}/all_metrics_validation.json'
plot_metric(metrics_train_file, metrics_val_file)

experiment_dir = '/datasets/models/parihaka_5a'
metrics_train_file = f'{experiment_dir}/all_metrics_train.json'
metrics_val_file = f'{experiment_dir}/all_metrics_validation.json'
plot_metric(metrics_train_file, metrics_val_file)

experiment_dir = '/datasets/models/parihaka_5b'
metrics_train_file = f'{experiment_dir}/all_metrics_train.json'
metrics_val_file = f'{experiment_dir}/all_metrics_validation.json'
plot_metric(metrics_train_file, metrics_val_file)

experiment_dir = '/datasets/models/parihaka_5c'
metrics_train_file = f'{experiment_dir}/all_metrics_train.json'
metrics_val_file = f'{experiment_dir}/all_metrics_validation.json'
plot_metric(metrics_train_file, metrics_val_file)

experiment_dir = '/datasets/models/parihaka_5d'
metrics_train_file = f'{experiment_dir}/all_metrics_train.json'
metrics_val_file = f'{experiment_dir}/all_metrics_validation.json'
plot_metric(metrics_train_file, metrics_val_file)

experiment_dir = '/datasets/models/parihaka_5e'
metrics_train_file = f'{experiment_dir}/all_metrics_train.json'
metrics_val_file = f'{experiment_dir}/all_metrics_validation.json'
plot_metric(metrics_train_file, metrics_val_file)

experiment_dir = '/datasets/models/parihaka_6a'
metrics_train_file = f'{experiment_dir}/all_metrics_train.json'
metrics_val_file = f'{experiment_dir}/all_metrics_validation.json'
plot_metric(metrics_train_file, metrics_val_file)

experiment_dir = '/datasets/models/parihaka_6b'
metrics_train_file = f'{experiment_dir}/all_metrics_train.json'
metrics_val_file = f'{experiment_dir}/all_metrics_validation.json'
plot_metric(metrics_train_file, metrics_val_file)

experiment_dir = '/datasets/models/parihaka_6c'
metrics_train_file = f'{experiment_dir}/all_metrics_train.json'
metrics_val_file = f'{experiment_dir}/all_metrics_validation.json'
plot_metric(metrics_train_file, metrics_val_file)

experiment_dir = '/datasets/models/parihaka_6d'
metrics_train_file = f'{experiment_dir}/all_metrics_train.json'
metrics_val_file = f'{experiment_dir}/all_metrics_validation.json'
plot_metric(metrics_train_file, metrics_val_file)

experiment_dir = '/datasets/models/parihaka_6e'
metrics_train_file = f'{experiment_dir}/all_metrics_train.json'
metrics_val_file = f'{experiment_dir}/all_metrics_validation.json'
plot_metric(metrics_train_file, metrics_val_file)
```
