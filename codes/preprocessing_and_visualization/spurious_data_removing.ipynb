```python
import os
import numpy as np
```

```python
def process_and_save_patches(data_file, label_file, boundaries_file):

    output_dir = data_file.split('/')[:-1]
    output_dir = '/'.join(output_dir)
    os.makedirs(output_dir, exist_ok=True)
    
    dataset_name = data_file.split('/')[-1].split('_')[:-1]
    dataset_name = '_'.join(dataset_name)
    print(f'Processing {dataset_name}')

    data = np.load(data_file, mmap_mode='r')  
    label = np.load(label_file, mmap_mode='r')
    boundaries = np.load(boundaries_file, mmap_mode='r')
    
    n_inlines = data.shape[0]
    n_crosslines = data.shape[1]
    n_timeslices = data.shape[2]

    first_inline = 0
    last_inline = n_inlines
    first_crossline = 0
    last_crossline = n_crosslines
    first_timeslice = 0
    last_timeslice = n_timeslices

    if 'f3' in dataset_name:
        first_inline = 30
        last_inline = 591
        first_crossline = 3
        last_crossline = 926
        first_timeslice = 70

    if 'penobscot' in dataset_name:
        first_inline = 70
        last_inline = 529

    print("Original data shape: ", data.shape)
    data = data[first_inline:last_inline, first_crossline:last_crossline, first_timeslice:last_timeslice]
    label = label[first_inline:last_inline, first_crossline:last_crossline, first_timeslice:last_timeslice]
    boundaries = boundaries[first_inline:last_inline, first_crossline:last_crossline, first_timeslice:last_timeslice]
    print("Processed data shape: ", data.shape)

    np.save(f'{output_dir}/{dataset_name}_dataset_cropped.npy', data)
    np.save(f'{output_dir}/{dataset_name}_label_cropped.npy', label)
    np.save(f'{output_dir}/{dataset_name}_boundaries_cropped.npy', boundaries)
        
```

```python
data_file = '/datasets/public_real_data/f3_silva_dataset.npy'
label_file = '/datasets/public_real_data/f3_silva_label.npy'
boundaries_file = '/datasets/public_real_data/f3_silva_boundaries.npy'

process_and_save_patches(data_file, label_file, boundaries_file)

data_file = '/datasets/public_real_data/penobscot_dataset.npy'
label_file = '/datasets/public_real_data/penobscot_label.npy'
boundaries_file = '/datasets/public_real_data/penobscot_boundaries.npy'

process_and_save_patches(data_file, label_file, boundaries_file)
```
